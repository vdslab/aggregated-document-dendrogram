```python
import pandas as pd
import json

with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/TextRank2018.json', encoding='utf8')as f:
  TextRank = pd.read_json(f)
with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/SingleRank2018.json', encoding='utf8')as f:
  SingleRank = pd.read_json(f)
with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/PositionRank2018.json', encoding='utf8')as f:
  PositionRank = pd.read_json(f)
with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/TopicRank2018.json', encoding='utf8')as f:
  TopicRank = pd.read_json(f)
with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/MultipartiteRank2018.json', encoding='utf8')as f:
  MultipartiteRank = pd.read_json(f)
with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/tfidf2018.json', encoding='utf8')as f:
  tfidf = pd.read_json(f)
with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/okapi2018.json', encoding='utf8')as f:
  okapi = pd.read_json(f)
```

```python
obj = []
for i in range(len(MultipartiteRank)):
  obj.append({'プロジェクトID':int(MultipartiteRank.loc[i]['プロジェクトID']),
              'TextRank':TextRank.loc[i]['TextRank'],
              'SingleRank':SingleRank.loc[i]['SingleRank'],
              'PositionRank':PositionRank.loc[i]['PositionRank'],
              'TopicRank':TopicRank.loc[i]['TipicRank'],
              'MultipartiteRank':MultipartiteRank.loc[i]['MultipartiteRank'],
              'tfidf':tfidf.loc[i]['tfidf'],
              'okapi':okapi.loc[i]['okapi']})
  
with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/phrase2018.json', 'w', encoding='utf8')as f:
  json.dump(obj, f, indent=2, ensure_ascii=False)
```
