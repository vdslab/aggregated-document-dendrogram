```python
from gensim.models.doc2vec import Doc2Vec
from top2vec import Top2Vec

dmodel = Doc2Vec.load('/content/drive/MyDrive/Colab Notebooks/pre-judgit/visdata_doc2vec.model')
tmodel = Top2Vec.load('/content/drive/MyDrive/Colab Notebooks/pre-judgit/visdata_top2vec.model')
```

```python
import pandas as pd
import csv

with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/IEEE VIS papers 1990-2020 - Main dataset.csv', 'r')as f:
    df = pd.read_csv(f)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tag import pos_tag
stem = PorterStemmer()
lem = WordNetLemmatizer()

def lemmatize_sentence(sentence):
    lemmatizer = WordNetLemmatizer()
    lemmatized_sentence = []
    for word, tag in pos_tag(word_tokenize(sentence.lower())):
        if tag.startswith('NN'):
            pos = 'n'
        elif tag.startswith('VB'):
            pos = 'v'
        else:
            pos = 'a'
        lemmatized_sentence.append(lemmatizer.lemmatize(word, pos))
    return lemmatized_sentence
```

```python
visdata_dic = {}
for a, t in zip(df['Abstract'], df['Title']):
    try:
        visdata_dic[' '.join(lemmatize_sentence(a))] = t
    except:
        pass
```

```python
x = dmodel.dv[visdata_dic[tmodel.documents[0]]]
y = dmodel.dv[visdata_dic[tmodel.documents[1]]]
```

```python
import numpy as np
def calc_cos_similarity(x, y):
    return np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
```

```python
calc_cos_similarity(tmodel.document_vectors[0], tmodel.document_vectors[1])
```

```python
visdata_dic[tmodel.search_documents_by_vector(tmodel.document_vectors[0], num_docs = 1)[0][0]]
```

```python
d_top100 = []
t_top100 = []
for i, (x, d) in enumerate(zip(tmodel.document_vectors, tmodel.documents)):
    if i % 10 == 0 : print(f'running {i}')
    d_top100.append([x[0] for x in dmodel.dv.similar_by_key(visdata_dic[d], topn = 100)])

    tmp_sort = []
    for i, y in enumerate(tmodel.document_vectors):
        tmp_sort.append((visdata_dic[tmodel.documents[i]], calc_cos_similarity(x, y)))
    t_top100.append([x[0] for x in sorted(tmp_sort, key=lambda x: x[1], reverse = True)[1:101]])
```
