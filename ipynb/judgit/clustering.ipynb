```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
with open("/content/drive/MyDrive/Colab Notebooks/pre-judgit/word_vector.json", encoding='utf8')as f:
  word_df = pd.read_json(f)
```

```python
word_df["vector"]
```

```python
for array in word_df["vector"]:
  if len(array) != 100:
    print("not 100")
```

```python
tmp = word_df["vector"][0]
```

```python
tmp
```

```python
word_array = []
for array in word_df["vector"]:
  tmp = [float(x) for x in array]
  word_array.append(tmp)
```

```python
word_array
```

```python
for array in word_array:
  if len(array) != 100:
    print("not 100")
```

```python
pred = KMeans(n_clusters=10).fit_predict(word_array)
```

```python
pred
```

```python
word_df["cluster_id"]=pred
```

```python
kmeans_model = KMeans(n_clusters=10, verbose=1, random_state=42, n_jobs=-1)
```

```python
kmeans_model.fit(word_array)
```

```python
kmeans_model.labels_
```

```python
word_df["cluster_id"]=kmeans_model.labels_
```
