```python
df_r=df.reset_index()
```

```python
df_r.loc[0]['事業概要（5行程度以内。別添可）']
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/pre-judgit/database2020.json', 'w', encoding='utf8')as f:
  for i in range(len(df_r)):
    f.write(json.dumps({'事業名':df_r.loc[i]['事業名'], '事業概要':df_r.loc[i]['事業概要（5行程度以内。別添可）']}, ensure_ascii=False) + '\n')
```
