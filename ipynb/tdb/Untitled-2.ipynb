```python
import pandas as pd
with open('/Users/taikitanaka/vsc/test/linkedDataCCR202101.json') as f:
    df = pd.read_json(f)
```

```python
import json
obj = []
for i in range(len(df)):
    a = []
    try:
        for k, v in df.loc[i]['MultipartiteRank'][0].items():
            a.append({  'word':k,
                        'score':v})
    except:
        pass
    obj.append({'no':str(df.loc[i]['no']),
                'parent':df.loc[i]['parent'],
                'distance':df.loc[i]['distance'],
                'Title':df.loc[i]['Title'],
                'Abstract':df.loc[i]['Abstract'],
                'AuthorKeywords':df.loc[i]['AuthorKeywords'],
                'Conference':df.loc[i]['Conference'],
                'MultipartiteRank':a})
with open('/Users/taikitanaka/vsc/test/linkedDataCCR202101.json', 'w', encoding='utf8')as f:
  json.dump(obj, f, indent = 2, ensure_ascii=False)
```
